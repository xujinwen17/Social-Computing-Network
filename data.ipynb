```python
import pandas as pd
```

```python
path1 = "data_by_artist.csv"
path2 = "influence_data.csv"
```

```python
df2 = pd.read_csv(path2)
df2
```

```python
artist2 = df2[['follower_id','follower_name', 'follower_main_genre', 'follower_active_start']]
artist2
```
